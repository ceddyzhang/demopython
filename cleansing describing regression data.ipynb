# Split training  data and test data

```python
from sklearn.model_selection import train_test_split
import pandas

#take only numeric data, which is only horesepower and highway-mpg
usedcars_clean_numeric_df = usedcars_clean_regression_df.loc[:,['symboling'
                                                                ,'normalized-losses'
                                                                ,'curb-weight'
                                                                ,'engine-size'
                                                                ,'horsepower'
                                                                ,'highway-mpg'
                                                                ,'price']].dropna(axis=0,how='any').astype('float')._get_numeric_data()


x_data_df = usedcars_clean_numeric_df.drop(['price'],axis=1)
y_data_df = usedcars_clean_numeric_df.loc[:,['price']]

x_data_df_train,x_data_df_test,y_data_df_train,y_data_df_test = train_test_split(x_data_df,y_data_df,test_size = 0.1,random_state=1)
x_data_df_train
```

### Fit using training data, evaluate using test data

```python
from sklearn.linear_model import LinearRegression
lre = LinearRegression()
lre.fit(x_data_df_train[['horsepower']],y_data_df_train)
#contrast R2
print('R2 of training data is '
      ,lre.score(x_data_df_train[['horsepower']],y_data_df_train)
      ,', R2 of test data is '
      ,lre.score(x_data_df_test[['horsepower']],y_data_df_test))
```

### Cross validation splits data into many folds. Some folds for training, others for testing, and then switch until each fold has been used for both training and test

```python
from sklearn.model_selection import cross_val_score

r2_cross = cross_val_score(lre,x_data_df[['horsepower']],y_data_df,cv=3)

print('Model lre s R2 on the 3 data folds are',r2_cross)
print('mean of r2_cross is ',r2_cross.mean(),', std dev is ',r2_cross.std())
```

# Ridge regression and alpha

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge

pr = PolynomialFeatures(degree=2)
x_data_df_poly_train = pr.fit_transform(x_data_df_train)
x_data_df_poly_test = pr.fit_transform(x_data_df_test)

feature_names = pr.get_feature_names_out(input_features = x_data_df_train.columns)

print('There are ',len(feature_names),' (6 orig features, 6C2 = 15 combinations of 2 or cross features, 6 square features, 1 constant) columns, which are ',feature_names)

#set and fit ridge model
ridgemodel = Ridge(alpha = 1)
ridgemodel.fit(x_data_df_poly_train,y_data_df_train)
yhat_df_ridge = ridgemodel.predict(x_data_df_poly_test)
```

# Grid search

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge

parameters_grid_search = [{'alpha':[0.01,0.1,1,10]}]

test_ridge = Ridge()
grid_1 = GridSearchCV(test_ridge
                      , parameters_grid_search
                      ,cv = 4)
grid_1.fit(x_data_df_poly_train,y_data_df_train)

best_ridge = grid_1.best_estimator_
print('the best alpha is ',best_ridge)

#test using score
best_ridge.score(x_data_df_poly_test,y_data_df_test)
```
